```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd


# cria o dataset
df = pd.read_csv('dados/Pokemon.csv')


# Crie dados aleatórios para x e y
x = np.array(df['Defense'])
y = np.array(df['Attack'])

# Ajuste um modelo de regressão linear aos dados
model = LinearRegression().fit(x.reshape(-1, 1), y)

# Calcule as previsões do modelo para valores de x
y_pred = model.predict(x.reshape(-1, 1))

# Plote os dados de treinamento e a linha de regressão
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.show()
```
